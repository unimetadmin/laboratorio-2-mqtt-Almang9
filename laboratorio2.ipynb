```python
import plotly.graph_objects as go
import sys
import paho.mqtt.client
import json

```

```python
contador_personas_data = []
f = go.FigureWidget()
f
f.layout.title = 'Contador Personas'
f.add_scatter(y=contador_personas_data)
```

```python
def on_connect(client, userdata, flags, rc):
	print('connected (%s)' % client._client_id)
	client.subscribe(topic='Casa/Sala/contador_personas', qos=0)
def on_message(client, userdata, message):
    print('------------------------------')
    print('topic: %s' % message.topic)
    print('payload: %s' % message.payload)
    print('qos: %d' % message.qos)
    resp =json.loads(message.payload)
    scatter = f.data[0]
    contador_personas_data.append(int(resp["Cantidad Personas"]))
    print(contador_personas_data)
    scatter.y = contador_personas_data

client = paho.mqtt.client.Client(client_id='contador-subs', clean_session=False)
client.on_connect = on_connect
client.on_message = on_message
client.connect('localhost',port=1883)
client.loop_forever()  
```
